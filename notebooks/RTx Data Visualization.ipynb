```python
def plot_two_col_scatter(col1, col2, df):
    plt.scatter(df[col1][df["did_the_patient_receive_pm"==1]],
           df[col2][df["did_the_patient_receive_pm"==1]],
           marker='D',
           color='red',
           label='RTx=1')
    plt.scatter(df[col1][df["did_the_patient_receive_pm"==0]],
           df[col2][df["did_the_patient_receive_pm"==0]],
           marker='o',
           color='blue',
           label='RTx=0')
```

```python
def plot_idx(col_idx):
    try:
        print("-"*100)
    #     plt.figure(figsize=(fig_size ,fig_size ))
        col_idx = int(col_idx)
        col_name = list(df.columns)[col_idx]
        col_type = col_types[col_name]
        print(f"Plotting column at index {col_idx}: {col_name} of type {col_type}...\n")
        pos_df = df.loc[df["did_the_patient_receive_pm"] == 1, col_name]
        neg_df = df.loc[df["did_the_patient_receive_pm"] == 0, col_name]
        plt.title(f"{col_name} ({col_type})" )

        if col_type == "enum": # Plot overlapping histograms
            bins = np.linspace(min(min(pos_df), min(neg_df)), max(max(pos_df), max(neg_df)), 20)
            plt.hist([neg_df, pos_df], bins, label=['RTx=0', 'RTx=1'], color=["blue", "orange"])
        #         plt.hist(pos_df, bins, label='RTx=1', color="orange")
            plt.legend(loc='best')
            plt.show()
            ax = plt.axes()
#             contingency_table = pd.crosstab(df["did_the_patient_receive_pm"], df[col_name], margins=False, dropna=False)
# #             percent = pd.crosstab(df["did_the_patient_receive_pm"], df[col_name], margins=False, normalize='columns').values
#             sns.heatmap(contingency_table,fmt=".0f",ax=ax, cmap='coolwarm', annot=True)
            contingency_table = pd.crosstab(df["did_the_patient_receive_pm"], df[col_name], margins=False, dropna=False)
#             percent = pd.crosstab(df["did_the_patient_receive_pm"], df[col_name], margins=False, normalize='columns').values
#             str_percent = []
#             for row in percent:
#                 temp = []
#                 for v in row:
#                     temp.append(str(v))
#                 str_percent.append(temp)
            sns.heatmap(contingency_table,fmt=".0f",ax=ax, cmap='coolwarm', annot=True)
            plt.show()
        else:
            f1 = plt.figure(1)
            plt.scatter(neg_df, [0] * len(neg_df), label='RTx=0', c="blue", alpha=0.1)
            plt.scatter(pos_df, [1] * len(pos_df), label='RTx=1', c="orange", alpha=0.1)
            plt.legend(loc='best')
            f1.show()
            f2 = plt.figure(2)
            sns.distplot(neg_df,  kde=False, label='RTx=0')
            sns.distplot(pos_df,  kde=False,label='RTx=1')
#             fig, ax = plt.subplots()
#             ax.hist(x, density=False, color='blue', alpha=0.5)
#             ax.hist(y, density=False, color='orange', alpha=0.5)
#             ax.set(title=f'Histogram of {col_name}')
            f2.show()
    except Exception as e:
#         print(f"Skipped plotting {col_name}")
        raise e
    plt.show()
```

```python
for col_idx in range(1, len(df)):
    try:
        plot_idx(col_idx)
    except:
        continue
```

```python
df_preds_auc = pd.read_csv("D:\YifuChen\Jan20-2022\H2ODAI\dai-1.10.1\data\Performance\AUC_feb10.csv")
df_preds_logloss = pd.read_csv("D:\YifuChen\Jan20-2022\H2ODAI\dai-1.10.1\data\Performance\LogLoss_tuned_feb.csv")
```

```python
def plot_preds(df_preds, title):
    pos_preds = df_preds.loc[df_preds["did_the_patient_receive_pm"] == 1, "did_the_patient_receive_pm.1"]
    neg_preds = df_preds.loc[df_preds["did_the_patient_receive_pm"] == 0, "did_the_patient_receive_pm.1"]
    plt.title(title)
    plt.scatter(pos_preds, [1] * len(pos_preds), label='RTx=1', alpha=0.05, c="orange")
    plt.scatter(neg_preds, [0] * len(neg_preds), label='RTx=0', alpha=0.05, c="blue")
    plt.legend(loc='best')
    plt.show()
```

```python
plot_preds(df_preds_logloss, "LogLoss_tuned")
```
